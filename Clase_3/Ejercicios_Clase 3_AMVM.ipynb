#### 1. Defina una secuencia de números aleatorios de longitud l, ubique el cero en la posición 2 en un color diferente. ¿Por qué es importante ubicar el cero en una secuencia? ¿Cuántas muestras se tomaron?

```python
def seq_rep(x_n,zero):
    zero=int(zero)
    N=len(x_n)
    n = np.arange(N)
    delta = np.zeros(N) # Vector de ceros
    delta[zero] = 1*x_n[zero]     # Se pone de color distintivo el 0 de la secuencia
    x_n[zero]=0
    plt.figure(figsize=(15,9))# tamaño de la figura
    plt.title('Representacion de secuencia')
    plt.stem(n-zero, x_n, 'b')
    plt.stem(n-zero, delta,'r',label='Cero de la secuencia')#pinta diferenciado el 0 en la representación
    plt.xlabel('Numero de muestras')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
l=10
Xn=np.random.randn(l)
posicion_0=2
seq_rep(Xn,posicion_0)
```

#### 2. Dadas las siguientes gráficas, defina cual es una secuencia causal, anticausal. ¿Por qué? 

```python
# Gráfica 1
Xn=np.array([1, 2, 3, 4])
posicion_0=0
seq_rep(Xn,posicion_0)
```
