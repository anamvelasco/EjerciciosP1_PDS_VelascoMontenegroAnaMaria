#### 1. Defina una señal con frecuencia = 10, discreticela y desplacela -5 unidades. Posteriormente desplacela 5 unidades. ¿Qué puede concluir de las dos gráficas?

```python
frecuencia=10
periodo = 1 / frecuencia
t = np.arange(0, 2*periodo, 0.01)
y = np.sin(2*np.pi*frecuencia*t)

# Discretizar la señal
n = np.arange(len(t))
y_discreta = np.sin(2*np.pi*frecuencia*n*periodo/len(t))

# Desplazar la señal periodica**
desplazamiento = -5
y_desplazada = np.roll(y_discreta, desplazamiento)

# Graficar señal original y desplazada
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
ax1.stem(n, y_discreta, use_line_collection=True)
ax1.set_title("Señal original")
ax1.set_xlabel("Muestras")
ax1.set_ylabel("Amplitud")

ax2.stem(n, y_desplazada, use_line_collection=True)
ax2.set_title("Señal desplazada")
ax2.set_xlabel("Muestras")
ax2.set_ylabel("Amplitud")

plt.tight_layout()
```

#### 3. Realice una función de la primera diferencia para secuencias no periódicas. Defina L=25. Grafique. ¿Que se puede decir de la primera diferencia? ¿Tiene alguna aplicación en la vida real?

```python
def primera_diferencia(X_n):
    #solo secuencias no periódicas
    X_diff=np.zeros(len(X_n))
    for n in range(len(X_n)-1):
        X_diff[n+1]=X_n[n+1]-X_n[n]
    X_diff[0]=X_n[0]
    n=np.arange(len(X_n))
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
    ax1.stem(n, X_n, use_line_collection=True)
    ax1.set_title("Señal original")
    ax1.set_xlabel("Muestras")
    ax1.set_ylabel("Amplitud")

    ax2.stem(n,X_diff,use_line_collection=True)
    ax2.set_title("Señal diferenciada")
    ax2.set_xlabel("Muestras")
    ax2.set_ylabel("Amplitud")

    plt.tight_layout()
    return X_diff
```
