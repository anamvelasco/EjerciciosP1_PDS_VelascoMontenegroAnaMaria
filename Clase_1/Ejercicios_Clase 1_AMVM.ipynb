```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Defina una señal con su respectivo vector de tiempo. Grafique en una misma figura una señal analógica y digital. Observe sus diferencias. ¿Qué puede decir de estas? 

```python
t = np.linspace(0, 2*np.pi, 20) # Vector de tiempo
t2 = np.linspace(0, 2*np.pi, 100) # Vector de tiempo
signal = np.sin(t)+ np.cos(5*t)   # Señal digital
signal2 = np.sin(t2)+ np.cos(5*t2)   # Señal digital
```

```python
# Grafica de la señal analógica y digital
plt.figure()
plt.plot(t2, signal2, 'b', label = 'Señal analógica')
plt.stem(t, signal, 'r', label = 'Señal digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
